```python
from pyspark.sql import SparkSession, functions as F
from pyspark.sql.types import StructType, StructField, DoubleType, StringType, IntegerType, DateType, ArrayType, FloatType
```

```python
spark = SparkSession.builder \
.master("local[2]") \
.appName("Filter") \
.getOrCreate()
```

```python
# ! wget -O /home/train/datasets/Hotel_Reviews.csv.gz \
# https://github.com/erkansirin78/datasets/raw/master/Hotel_Reviews.csv.gz
```

# Read Data

```python
df = spark.read \
.format("csv") \
.option("header", True) \
.option("sep", ",") \
.option("inferSchema", True) \
.load("file:///home/train/datasets/Hotel_Reviews.csv.gz") \
.withColumn("Review_Date", F.to_date("Review_Date", "M/d/y")) \
.withColumn("Tags", F.split(F.col("Tags"), ",").cast(ArrayType(StringType())))
```

```python
df.limit(6).toPandas()
```

```python
df.where("TRIM(Reviewer_Nationality) == 'United Kingdom'").limit(5).toPandas()
```

```python
df.filter("Reviewer_Nationality LIKE '%United Kingdom%' AND Average_Score < 7.0").limit(5).toPandas()
```

```python
df.filter(F.col("Reviewer_Nationality").like('%United Kingdom%')).limit(5).toPandas()
```

```python
country = 'United Kingdom'
```

```python
df.where(f"TRIM(Reviewer_Nationality) == '{country}' AND Average_Score < 7.0").limit(5).toPandas()
```
